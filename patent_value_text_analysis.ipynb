This notebook analyzes 2022 patents from the KPSS dataset, merges firm identifiers, extracts abstracts from USPTO grant data, and studies how vocabulary usage relates to patent value.

## Parse Patent Grant XML and Extract Abstracts

```python
# Load XML parsing tools and patent grant archive path
from zipfile import ZipFile
from lxml import etree
from lxml.etree import parse, XMLParser

# Set path to location where downloaded ipg data is located
path_to_xml = "ipg221213.zip"
```

```python
# Open the archived XML file
with ZipFile(path_to_xml, 'r') as z:
    archive_to_open = z.open(z.namelist()[0],'r')
    with archive_to_open as z2:
        contents = z2.read()
    
print(type(contents))
print(contents[:100])
```

```python
# Split the raw USPTO XML into individual records
formatted_data = []

for rec in contents.decode('utf8').split("<?xml"):
    if rec == '': # Skip empty lines
        continue
    reformatted_rec = "<?xml" + rec # add back the '<?xml'
    formatted_data.append(reformatted_rec.encode('utf8'))  
    
# print(formatted_data[0][:100]) 
# print(formatted_data[5][:100])
print(f"The number of records in formatted_data is: {len(formatted_data)}") # check length
```

```python
# Parse each XML record into an lxml tree
forest = [] 
for xml in formatted_data: 
    tree = etree.fromstring(xml) 
    forest.append(tree)          
```

```python
# Extract patent numbers and abstract text
# Iterate through the data and extract the patent number and abstract
import pandas as pd
from lxml import etree

patent_records = []

for tree in forest:  
    # extract patent number
    patent_number_tag = tree.find('.//publication-reference/document-id/doc-number')
    patent_number = patent_number_tag.text.strip() if patent_number_tag is not None and patent_number_tag.text else None

    # extract abstract 
    abstract_tag = tree.find('.//abstract')
    abstract_text = None
    if abstract_tag is not None:
        text = ''.join(abstract_tag.itertext()).strip()
        if text:
            abstract_text = text

    # append to list
    if patent_number:  
        patent_records.append({'patent_number': patent_number, 'abstract': abstract_text})
```

```python
# Store extracted patent abstracts in a dataframe
patent_abstracts = pd.DataFrame(patent_records)
patent_abstracts
```

## Build a Document-Term Matrix from Patent Abstracts

```python
# Drop missing abstracts and align merge keys
patent_abstracts_good = patent_abstracts.dropna(subset=['abstract']).reset_index(drop=True) 

patent_abstracts_good = patent_abstracts_good.rename(columns={'patent_number': 'patent_num'})
sub2['patent_num'] = sub2['patent_num'].astype(str)
# merge with the patent data 
df_good = patent_abstracts_good.merge(sub2, on='patent_num', how='left')

# print(df_good.head())
# print(df_good.info())
```

```python
# Custom tokenizer for patent abstracts
import spacy
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

nlp = spacy.load('en_core_web_lg')
stops = stopwords.words('english')

def parse_document(doc):
    doc = nlp(doc)
    words = []  
    
    for sent in doc.sents:  
        for word in sent:
            word_lower = word.text.lower()  # use lowercase
            if word.is_alpha and len(word_lower) >= 3:   # include only those tokens that are all letters & require tokens be at least 3 characters long
                if word_lower in stops: # Exclude stop words
                    continue
                words.append(word_lower)
                
    return words
```

```python
# Build the document-term matrix
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

vec = CountVectorizer(tokenizer=parse_document, ngram_range=(1,2), max_features=1000) # Allow for single words and bigrams & Restrict the matrix to the 1,000 most common words
dtm = vec.fit_transform(df_good['abstract']) 
```

```python
import numpy as np

dtm_dense_df = pd.DataFrame(dtm.todense(),columns=vec.get_feature_names())
# print(f"Number of columns in dtm_dense_df: {dtm_dense_df.shape}")

total_words = dtm_dense_df.sum(axis=0)
top25 = total_words.nlargest(25)
print(f"Top 25 most frequently used words:\n{top25}")
```

## Correlate Terms with Patent Value

```python
# Keep rows with both abstract text and patent value
df_wd_value = df_good.dropna(subset=['xi_real']).reset_index(drop=True) 

# df_wd_value.info()
# df_wd_value.head(5)
```

```python
# Utility function for pairwise correlation
import numpy as np

def cal_cor(x, y):
    x = np.array(x)
    y = np.array(y)
    
    if np.std(x) == 0:
        return 0.0
    return np.corrcoef(x, y)[0, 1]
```

```python
# Calculate correlations between term counts and patent value
vec = CountVectorizer(tokenizer=parse_document, ngram_range=(1,2), max_features=1000)
dtm_wd_value = vec.fit_transform(df_wd_value['abstract']) 

dtm_wd_value_df = pd.DataFrame(dtm_wd_value.toarray(), columns=vec.get_feature_names())
xi_real_value = df_wd_value['xi_real'].values # xi_real array

# Iterate through each word and calculate the correlation
corr_dict = {}
for word in dtm_wd_value_df.columns:
    word_counts = dtm_wd_value_df[word].values
    corr = cal_cor(word_counts, xi_real_value)
    corr_dict[word] = corr

# Step3: Create a pandas Series with the final results, where the index of the series is the word.    
corr_series = pd.Series(corr_dict)
# corr_series
```

```python
# Sort correlations from most negative to most positive
corr_series_sorted = corr_series.sort_values() # Sort the series in ascending order

# most negative:
most_negative = corr_series_sorted.head(10)
print(f"Top 10 most negatively correlated words with xi_real:\n{most_negative}")
```

```python
# most positive:
most_positive = corr_series_sorted.tail(10).iloc[::-1]
print(f"Top 10 most positively correlated words with xi_real:\n{most_positive}")
```

## Interpretation
The strongest positive term-level relationship is `test`, with other positive terms such as `processors`, `processes`, and `chip` also concentrated in technical invention language. The most negative terms are weaker in magnitude, suggesting that abstract text contains some signal, but that patent value is also driven by broader firm and market factors.
